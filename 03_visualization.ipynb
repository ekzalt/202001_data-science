```python
adf = pd.DataFrame({
  'country': ['Russia', 'Belarus', 'Ukraine'],
  'population': [143.5, 9.5, 45.5],
  'square': [17125191, 207600, 603628],
})
adf['density'] = adf['population'] / adf['square'] * 1_000_000

adf['population'].value_counts().plot.bar()
# (adf['population'].value_counts() / len(adf)).plot.bar()

adf
```

```python
bdf = pd.DataFrame({
  'sex': [1, 0, 1, 1, 0],
  'weight': [70, 50, 80, 90, 60],
  'age': [20, 20, 30, 40, 30],
  'salary': [1500, 1000, 2000, 1000, 1500],
})

sb.violinplot(x='age', y='salary', hue='sex', split=True, data=bdf)

bdf
```

```python
bdf_columns = ['sex', 'weight', 'age', 'salary']
sb.pairplot(bdf[bdf_columns])
```
